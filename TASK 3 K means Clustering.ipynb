### Creating the Matrix of Features

```python
X = data.iloc[:, 0:4].values
```

```python
X
```

## We use the Elbow Method to determine the Optimal Number of Clusters 

```python
# Using the Elbow Method to find the optimal number of clusters 
from sklearn.cluster import KMeans
wcss = []   # wcss - within cluster sum of squares
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)  
plt.plot(range(1,11), wcss)  # Plotting the graph
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

## As shown in the graph above it forms an Elbow, the optimal number of clusters are where the Elbow occurs. In this case the Elbow occurs at 3, therefore, the optimal number of clusters in this dataset is '3'. 

```python
# Applying K-Means Clustering to the Iris dataset
kmeans = KMeans(n_clusters = 3, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
y_kmeans = kmeans.fit_predict(X)
```

```python
y_kmeans
```

```python
y_kmeans.shape
```

## Visualising the Clusters using Scatter Plot

```python
plt.figure(figsize=(10,7))
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 100, c = 'blue', label = 'Iris-setosa ')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 100, c = 'red', label = 'Iris-virginica')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 100, c = 'green', label = 'Iris-versicolor')

plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroids')

plt.legend()
plt.show()
```
